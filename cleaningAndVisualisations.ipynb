```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("vgsales_dirty.csv")
print(data.head(5))
```

```python
sns.boxplot(x=data["Global_Sales"])

plt.title("Boxplot of Global Video Game Sales")
plt.xlabel("Global Sales (Millions of Units)")

plt.show()
```

```python
sns.histplot(data["Global_Sales"], bins=50)

plt.title("Distribution of Global Video Game Sales")
plt.xlabel("Global Sales (Millions)")
plt.ylabel("Number of Games")

plt.show()

```

```python
sns.histplot(data["Year"], bins=40)

plt.title("Distribution of Game Release Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Games")

plt.show()
```

```python
print(data.info())
print(data.describe())
print(data.shape)
print(data.isna().sum())
```

```python
data.columns = (
    data.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

data.head()
```

```python
data["publisher"] = data["publisher"].replace("nan", "Unknown")
data["publisher"] = data["publisher"].fillna("Unknown")
print(data["publisher"].isna().sum())
```

```python
data = data.dropna(subset=["year"])
print("After dropping missing years:", data.shape)
```

```python
data.duplicated().sum()
data = data.drop_duplicates()
```

```python
sales_cols = ["na_sales","eu_sales","jp_sales","other_sales","global_sales"]
for col in sales_cols:
    data[col] = pd.to_numeric(data[col], errors="coerce")
```

```python
# data.to_csv("vgsales_cleaned.csv", index=False)

print("File saved as vgsales_cleaned.csv")
```

# Visualisations

## Loading the dataset

```python
cleanData = pd.read_csv("vgsales_cleaned.csv")
display(cleanData.head(10))
```

**Data description:**<br>
- rank - rank
- name - categorical nominal
- platform - categorical nominal
- year - numerical discrete
- genre - categorical nominal
- publisher - categorical nominal
- na_sales - numerical continuous
- eu_sales - numerical continuous
- jp_sales - numerical continuous
- other_sales - numerical continuous
- global_sales - numerical continuous

## Top Platforms by Global Sales

```python
platformSales = (cleanData.groupby("platform")["global_sales"].sum().sort_values(ascending = False).head(10).reset_index())

sns.barplot(data = platformSales, x = "global_sales", y = "platform")
plt.title("Top 10 Platforms by Global Sales")
plt.xlabel("Global Sales (millions)")
plt.ylabel("Platform")
plt.tight_layout()
plt.show()
```

## Sales over Time

```python
yearlySales = cleanData.groupby("year")["global_sales"].sum().reset_index()

sns.lineplot(data = yearlySales, x = "year", y = "global_sales", marker = "o")
plt.title("Global Sales Over Time")
plt.xlabel("Year")
plt.ylabel("Global Sales (millions)")
plt.tight_layout()
plt.show()
```

## Genre performance

```python
sns.boxplot(data = cleanData, x = "genre", y = "global_sales")
plt.xticks(rotation = 45)
plt.title("Global Sales Distribution by Genre")
plt.tight_layout()
plt.show()
```

## Regional sales by Genre

```python
regionByGenre = (cleanData.groupby("genre")[["na_sales", "eu_sales", "jp_sales", "other_sales"]].sum())

sns.heatmap(regionByGenre, annot = True, fmt = ".0f", cmap = "coolwarm")
plt.title("Regional Sales by Genre")
plt.tight_layout()
plt.show()
```

## Region Correlations

```python
sns.pairplot(cleanData[["na_sales", "eu_sales", "jp_sales", "other_sales", "global_sales"]])
plt.tight_layout()
plt.show()
```
